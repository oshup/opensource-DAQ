# Value Display GUI

```python
import tkinter as tk
import serial
import time
import numpy as np
import threading
import continuous_threading
from tkinter import * 
from matplotlib.figure import Figure
from matplotlib.backends.backend_tkagg import (FigureCanvasTkAgg, 
NavigationToolbar2Tk)


ser = serial.Serial('COM5', 9600)
time.sleep(2)



val1 = 0

index = []

def readserial():
    global val1
    ser_bytes = ser.readline()
    ser_bytes = ser_bytes.decode("utf-8")
    print(ser_bytes.rstrip())
    val1 = ser_bytes
    index.append(val1)
    
    if len(index) == 1:
        display1 = tk.Label(window,text=index[0]).place(x=10,y=30)
   
        
    if len(index) == 2:
        index.clear()
    
    time.sleep(0.5)

t1 = continuous_threading.PeriodicThread(0.5, readserial)

window = Tk()


window.geometry("300x150")
window.title('Temperature Value GUI')
w = tk.Button(window,text=" Start", command = t1.start, height= 1, width=5).place(x=10,y=1)
w1 = tk.Button(window,text="Stop", command = t1.stop,height= 1, width=5).place(x=10,y=80)
w2 = tk.Label(window,text='Temperature (C)').place(x=65,y=30)


window.mainloop()
```

# Graph GUI

```python
import tkinter as tk
import serial
import time
import numpy as np
import threading
import continuous_threading
from tkinter import * 
from matplotlib.figure import Figure
from matplotlib.backends.backend_tkagg import (FigureCanvasTkAgg, 
NavigationToolbar2Tk)

ser = serial.Serial('COM5', 9600)
time.sleep(2)

data =[] 
for i in range(40):
    b = ser.readline()      
    string_n = b.decode()   
    string = string_n.rstrip() 
    print(string)
    data.append(string)            
    time.sleep(0.1)           
ser.close()



time = range(0,40,1)
    
def plot():
  
   
    fig = Figure(figsize = (20, 20),
                 dpi = 80)
    
    x = time
    y = data
  

    plot1 = fig.add_subplot(111)
    plot1.set_title('Temperature Graph')
    plot1.set_xlabel('time(s)')
    plot1.set_ylabel('Temperature (C)')
    
  
    plot1.plot(x,y)
  
    
    canvas = FigureCanvasTkAgg(fig,
                               master = window)  
    canvas.draw()
    canvas.get_tk_widget().pack()
    toolbar = NavigationToolbar2Tk(canvas,
                                   window)
    toolbar.update()
      
window = Tk()   
window.title('Temperature Graph GUI')
window.geometry("800x800")
plot_button = Button(master = window, command = plot,
                     height = 2, 
                     width = 20,
                     text = "On/Off")

plot_button.pack()


window.mainloop()

```
